```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
import numpy as np
import matplotlib.pyplot as plt

# Read the data
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
#df.head()
#df['VIN (1-10)']
#df.columns #['VIN (1-10)', 'County', 'City', 'State', 'Postal Code', 'Model Year','Make', 'Model', 'Electric Vehicle Type', 'Clean Alternative Fuel Vehicle (CAFV) Eligibility', 'Electric Range','Base MSRP', 'Legislative District', 'DOL Vehicle ID','Vehicle Location', 'Electric Utility', '2020 Census Tract']
# 1. Cars and their types that do not qualify for clean alternative fuel vehicle
non_clean_fuel_cars = df[df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].str.lower() != 'Clean Alternative Fuel Vehicle (CAFV) Eligibility']
non_clean_fuel_cars = non_clean_fuel_cars[['Make', 'Electric Vehicle Type']]
print("Cars and Their Types that Do Not Qualify for Clean Alternative Fuel Vehicle:")
print(non_clean_fuel_cars)
# 2. TESLA cars with the model year, and model type made in Bothell City
tesla_bothell_cars = df[(df['Make'] == 'TESLA') & (df['City'] == 'Bothell')]
tesla_bothell_cars = tesla_bothell_cars[['Model Year', 'Model']]
print("\nTESLA Cars with Model Year and Model Type Made in Bothell City:")
print(tesla_bothell_cars)
# 3. Cars with electric range > 100 and made after 2015
electric_cars = df[(df['Electric Range'] > 100) & (df['Model Year'] > 2015)]
print("\nCars with Electric Range > 100 and Made After 2015:")
print(electric_cars[['Make', 'Model Year', 'Electric Range']])

# 4. Distribution plot between city and electric vehicle type
city_vehicle_type_distribution = df.groupby(['City', 'Electric Vehicle Type']).size().unstack()
plt.figure(figsize=(50, 50))
city_vehicle_type_distribution.plot(kind='bar', stacked=True)
plt.xlabel('City')
plt.ylabel('Count')
plt.title('Distribution of City and Electric Vehicle Type')
plt.legend(loc='upper right')
plt.show()
```
