```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the meteorites data into a DataFrame
df = pd.read_csv('earth_meteorites.csv')

# Convert the 'year' column to datetime type
df['year'] = pd.to_datetime(df['year'], errors='coerce')

# Get all the Earth meteorites that fell before the year 2000
earth_meteorites_before_2000 = df[(df['reclat'].notnull()) & (df['year'].dt.year < 2000)]

# Plot the distribution of Earth meteorites by year
earth_meteorites_before_2000['year'].dt.year.value_counts().sort_index().plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Distribution of Earth Meteorites (Before 2000)')
plt.show()

# Get all the Earth meteorites' coordinates that fell before the year 1970
earth_meteorites_before_1970 = df[(df['reclat'].notnull()) & (df['year'].dt.year < 1970)]

# Plot the distribution of Earth meteorites by latitude and longitude
plt.scatter(earth_meteorites_before_1970['reclong'], earth_meteorites_before_1970['reclat'], s=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Coordinates of Earth Meteorites (Before 1970)')
plt.show()

# Assuming mass is in kg, get all Earth meteorites with a mass greater than 10000 kg
earth_meteorites_mass_gt_10000 = df[(df['reclat'].notnull()) & (df['mass'] > 10000)]

# Plot the distribution of Earth meteorites by mass
earth_meteorites_mass_gt_10000['mass'].plot(kind='hist', bins=20)
plt.xlabel('Mass')
plt.ylabel('Count')
plt.title('Distribution of Earth Meteorites with Mass > 10000 kg')
plt.show()
```
